```python
def to_black_and_white(image) :
    shape = image.shape
    dim1 = shape[0]
    dim2 = shape[1]
    color = np.zeros((dim1,dim2,3))
    for i in range(dim1):
        for j in range(dim2):
            if(image[i][j] == 255):
                color[i][j][0] = 255
                color[i][j][1] = 255
                color[i][j][2] = 255
    return color
```
