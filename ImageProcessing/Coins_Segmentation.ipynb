```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
image = cv2.imread("/home/guru/Desktop/OpenCV/ImageProcessing/water_coins.jpg",0)
```

```python
plt.imshow(image,cmap = "gray")
```

```python
ret,thresh = cv2.threshold(image,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
kernel = np.ones((7,7),np.uint8)
```

```python
erode = cv2.erode(thresh,kernel,iterations = 4)
```

```python
plt.imshow(erode,cmap = "gray")
```

```python
outputs = cv2.connectedComponentsWithStats(erode,cv2.CV_64F)
```

```python
#Gives no of circles
outputs[0]-1
```

```python
plt.imshow(outputs[1],cmap = "gray")
```

```python
hist = cv2.equalizeHist(image)
```
