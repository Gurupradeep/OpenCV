```python
kernel = np.ones((3,3),np.uint8)
```

```python
#Closing to remove black dots
closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
```

```python
plt.imshow(closing)
```

```python
# 4 is connectivity, closing is grayscale image
#Output is tuple
# Refer here https://docs.opencv.org/3.1.0/d3/dc0/group__imgproc__shape.html#gae57b028a2b2ca327227c2399a9d53241

output = cv2.connectedComponentsWithStats(closing,4,cv2.CV_32S)
```

```python
#Gives no of labels(connected components) 
# Note : Includes background also
output[0]
```

```python
#Gives labelled image
plt.imshow(output[1])
```

```python
#statistics output for each label
# Last column denotes the total area (in pixels) of the connected component. 
# For remaining refer https://docs.opencv.org/3.1.0/d3/dc0/group__imgproc__shape.html#gac7099124c0390051c6970a987e7dc5c5
output[2]
```

```python
#Getting the required things
components = output[0]
labelled_image = output[1]
stats = output[2]
```

```python
# Gives the total area of each connected component
sizes = stats[:, -1]
```

```python
#Gives the connected component with maximum area
max_label = 1
max_size = sizes[1]
for i in range(2, components):
    if sizes[i] > max_size:
        max_label = i
        max_size = sizes[i]
img2 = np.zeros(labelled_image.shape)
img2[labelled_image == max_label] = 255
```
