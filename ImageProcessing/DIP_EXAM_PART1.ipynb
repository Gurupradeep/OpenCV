```python
retval,thresh = cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh,cmap = "gray")
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations = 2)
```

```python
plt.imshow(opening,cmap = "gray")
```

```python
outputs = cv2.connectedComponentsWithStats(opening,cv2.CV_64F)
```

```python
connected = outputs[0]
```
