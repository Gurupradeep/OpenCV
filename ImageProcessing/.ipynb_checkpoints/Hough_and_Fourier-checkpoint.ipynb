```python
image = cv2.imread("/home/guru/Desktop/OpenCV/ImageProcessing/Edge2.jpeg",0)
```

```python
f = np.fft.fft2(image)
```

```python
fshift = np.fft.fftshift(f)
```

```python
magnitude_spectrum = 20*np.log(np.abs(fshift))
```

```python
plt.imshow(magnitude_spectrum,cmap = "gray")
```

```python
#Applying high pass filter in frequency domain
#Making the center pixels value zero

rows,cols = image.shape

#Getting centre row and column 
crow = rows/2
ccol = cols/2

fshift[crow-30:crow+30,ccol-30:ccol+30] = 0

#Now applying inverse fourier transformation

i_fshift = np.fft.ifftshift(fshift)
new_image = np.fft.ifft2(i_fshift)
new_image = np.abs(new_image)
```

```python
plt.imshow(new_image,cmap ="gray")
```

```python
f = np.fft.fft2(image)
fshift = np.fft.fftshift(f)
```

```python
#Applying low pass filter
#Making all pixels other than the ones near centre zero

rows,cols = image.shape
mask = np.zeros((rows,cols))

#Getting centre row and column

crow = rows/2
ccol = cols/2

#Making mask
mask[crow-30:crow+30,ccol-30:ccol+30] = 1

f_shift = fshift*mask

i_fshift = np.fft.ifftshift(f_shift)
smooth_image = np.fft.ifft2(i_fshift)
smooth_image = np.abs(smooth_image)
```

```python
plt.imshow(smooth_image,cmap = "gray")
```

```python
#Making a binary mask which has all zeros expect for the lines
# We multiply this with the output of the canny edge detection to get the actual lines in the image
mask = np.zeros(image.shape)
```

```python
for i in range(lines.shape[0]):
    for rho,theta in lines[i]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(mask,(x1,y1),(x2,y2),(255,255,255),1)
```
