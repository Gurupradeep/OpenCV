```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# TO FIND THE CENTRE OF USED TABLET

```python
image = cv2.imread("/home/guru/Desktop/OpenCV/ImageProcessing/image1.jpg",0)
```

```python
cimg = cv2.imread("/home/guru/Desktop/OpenCV/ImageProcessing/image1.jpg")
```

```python
plt.imshow(image,cmap = "gray")
```

```python
img = cv2.medianBlur(image,5)
```

```python
plt.imshow(img,cmap = "gray")
```

```python
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,40,
                            param1=50,param2=30,minRadius=22,maxRadius=50)
```

```python
circles = np.uint16(np.around(circles))
```

```python
circles[0,:]
```

```python
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
```

```python
plt.imshow(cimg)
```
